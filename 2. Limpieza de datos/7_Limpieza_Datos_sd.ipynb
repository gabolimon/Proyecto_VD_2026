```python
# CONVERTIR USER_ID EN STRING
df_sd['user_id'] = df_sd['user_id'].astype(str)

# ELIMNAR COLUMNA NIGTH_NOTIFICATIONS
df_sd.drop(columns=['night_notifications'], inplace=True)
```

```python
# CAMBIAR NOMBRE DE COLUMNAS
df_sd.rename(columns={
    'user_id'               : 'id_usuario',
    'year'                  : 'anio',
    'country'               : 'pais',
    'age_group'             : 'grupo_edad',
    'gender'                : 'genero',
    'platform'              : 'plataforma',
    'avg_sleep_hours'       : 'promedio_horas_sueno',
    'sleep_quality_score'   : 'puntaje_calidad_sueno',
    'late_night_usage_hours': 'horas_uso_nocturno',
    'fatigue_level'         : 'nivel_fatiga'
}, inplace=True)
```

```python
# VERIFICACIÓN
df_sd.info()
```

<h3>AGREGACIÓN DE NUEVOS ATRIBUTOS</h3>

```python
# CLASIFICADOR DE SUEÑO
bins = [0, 5, 7, 9, 12]
labels = ['Insuficiente', 'Aceptable', 'Optimo', 'Excesivo']
df_sd['clasificacion_sueno'] = pd.cut(df_sd['promedio_horas_sueno'], bins=bins, labels=labels, include_lowest=True)

# Categoriza las horas de sueño según los estándares clínicos de la National Sleep Foundation: menos de 5 horas es insuficiente, 7–9 es el rango óptimo. 
# Útil para identificar pacientes en riesgo por privación de sueño.

#FATIGA CATEGORIA
bins = [0, 33, 66, 100]
labels = ['Leve', 'Moderada', 'Severa']
df_sd['fatiga_categoria'] = pd.cut(df_sd['nivel_fatiga'], bins=bins, labels=labels, include_lowest=True)

```
