```python
# CONVERTIR USER_ID EN STRING
df_mht['user_id'] = df_mht['user_id'].astype(str)
```

```python
# CAMBIAR NOMBRE DE COLUMNAS

df_mht.rename(columns={
    'user_id'           : 'id_usuario',
    'year'              : 'anio',
    'country'           : 'pais',
    'age_group'         : 'grupo_edad',
    'gender'            : 'genero',
    'platform'          : 'plataforma',
    'anxiety_score'     : 'puntaje_ansiedad',
    'depression_score'  : 'puntaje_depresion',
    'stress_level'      : 'nivel_estres',
    'loneliness_index'  : 'indice_soledad',
    'therapy_access'    : 'acceso_terapia',
    'medication_usage'  : 'uso_medicacion',
    'self_esteem_score' : 'puntaje_autoestima',
    'mental_health_risk': 'riesgo_salud_mental'
}, inplace=True)
```

```python
# VERIFICACIÓN
df_mht.info()
```

<h3>AGREGACIÓN DE NUEVOS ATRIBUTOS</h3>

```python
# INDICE BIENESTAR DIGITAL
df_mht['indice_bienestar_digital'] = 100 - ((df_mht['puntaje_ansiedad'] + df_mht['puntaje_depresion'] + df_mht['nivel_estres']) / 3).round(2)

# Resume en un solo número el estado psicologico del paciente. Valor alto equivale a mejor bienestar. 

# BRECHA DE AUTOESTIMA ANSIEDAD
df_mht['brecha_autoestima_ansiedad'] = (df_mht['puntaje_autoestima'] - df_mht['puntaje_ansiedad']).round(2)

# Una baja autoestima combinada con una alta ansiedad es un patron de riesgo, valores negativos indican que la ansiedad supera el autoestima. 

# APOYO CLINICO
df_mht['apoyo_clinico'] = (df_mht['acceso_terapia'].astype(int) + df_mht['uso_medicacion'].astype(int))
# 0 = sin apoyo, 1 = apoyo parcial, 2 = apoyo completo

# RIESGO ORDEN
orden = {'Low': 1, 'Medium': 2, 'High': 3}
df_mht['riesgo_orden'] = df_mht['riesgo_salud_mental'].map(orden)
```
