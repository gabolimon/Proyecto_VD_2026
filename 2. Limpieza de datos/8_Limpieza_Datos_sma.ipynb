<h3>AGREGACIÓN DE NUEVOS ATRIBUTOS</h3>

```python
# USO DE PANTALLA CATEGORIA
bins = [0, 3, 6, 9, 15]
labels = ['Bajo', 'Moderado', 'Alto', 'Severo']
df_smu['uso_pantalla_categoria'] = pd.cut(df_smu['horas_pantalla_diaria'], bins=bins, labels=labels)

# Categoriza las horas de pantalla en niveles clínicos. En medicina digital se considera uso problemático a partir de 6 horas diarias.

# DOOMSCROLLING CATEGORIA
bins = [0, 3, 6, 10]
labels = ['Leve', 'Moderado', 'Severo']
df_smu['doomscrolling_categoria'] = pd.cut(df_smu['frecuencia_doomscrolling'], bins=bins, labels=labels, include_lowest=True)

#Segmenta el doomscrolling en niveles.
```
