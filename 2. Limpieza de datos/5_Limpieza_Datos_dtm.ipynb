```python
ruta_parquet = Path(r"../1. Datasets/1. Dataset Original")

df_dtm = pd.read_csv(ruta_parquet/"dopamine_trigger_metrics.csv")
```

```python
# RANGO AÑO 
print("Min:", df_dtm['year'].min(), "| Max:", df_dtm['year'].max())
```

```python
# CONVERTIR USER_ID EN STRING
df_dtm['user_id'] = df_dtm['user_id'].astype(str)

#ELIMNAR COLUMAN CONTENT_REFRESH_FRECUENCY
df_dtm.drop(columns=['content_refresh_frequency'], inplace=True)
```

```python
# CAMBIAR NOMBRE DE COLUMNAS
df_dtm.rename(columns={
    'user_id'                          : 'id_usuario',
    'year'                             : 'anio',
    'country'                          : 'pais',
    'age_group'                        : 'grupo_edad',
    'gender'                           : 'genero',
    'platform'                         : 'plataforma',
    'engagement_rate'                  : 'tasa_interaccion',
    'short_video_consumption_hours'    : 'horas_consumo_video_corto',
    'dopamine_trigger_score'           : 'puntaje_activacion_dopamina',
    'instant_gratification_dependency' : 'dependencia_gratificacion_inmediata'
}, inplace=True)
```

```python
# VERIFICACIÓN
df_dtm.info()
```

<h3>AGREGACIÓN DE NUEVOS ATRIBUTOS</h3>

```python
# CONSUMO VIDEO CATEGORIA
bins = [0, 2, 5, 8, 10]
labels = ['Bajo', 'Moderado', 'Alto', 'Severo']
df_dtm['consumo_video_categoria'] = pd.cut(df_dtm['horas_consumo_video_corto'], bins=bins, labels=labels, include_lowest=True)

#Categoriza las horas de consumo de video corto en niveles. En contexto clínico más de 5 horas diarias de video corto se considera consumo problemático.

# TASA INTERACCION CATEGORIA
bins = [0, 33, 66, 100]
labels = ['Baja', 'Media', 'Alta']
df_dtm['tasa_interaccion_categoria'] = pd.cut(df_dtm['tasa_interaccion'], bins=bins, labels=labels, include_lowest=True)

#Agrupa la tasa de interacción en tercios.
```

```python
# RUTA DONDE SE GUARDARÁN LOS PARQUET
ruta_limpio = Path(r"../1. Datasets/3. Dataset Limpio-csv")

# EXPORTAR CADA DATASET COMO PARQUET
df_dtm.to_parquet(ruta_limpio/"dopamine_trigger_metrics.parquet", index=False)

# EXPORTAR CADA DATASET COMO CSV
df_dtm.to_csv(ruta_limpio/"dopamine_trigger_metrics.csv", index=False)
```
